```python
import pandas as pd

# Assuming df1 has 'Fecha' column and df2 has 'Date_Range' column

# Convert 'Fecha' column in df1 to datetime if it's not already
df1['Fecha'] = pd.to_datetime(df1['Fecha'])

# Convert 'Date_Range' column in df2 to datetime and split it into 'Start_Date' and 'End_Date'
df2['Date_Range'] = pd.to_datetime(df2['Date_Range'])
df2[['Start_Date', 'End_Date']] = df2['Date_Range'].str.split('-', expand=True).apply(lambda x: x.str.strip())

# Initialize an empty list to store the boolean masks for each pair of rows in df2
masks = []

# Iterate over the rows in df2, considering two consecutive rows at a time
for i in range(len(df2) - 1):
    # Get the start and end dates from the current and next rows of df2
    start_date = df2.loc[i, 'Start_Date']
    end_date = df2.loc[i + 1, 'End_Date']
    
    # Create a boolean mask for df1 based on whether each date is within the current date range
    mask = (df1['Fecha'] >= start_date) & (df1['Fecha'] <= end_date)
    
    # Append the mask to the list of masks
    masks.append(mask)

# Concatenate all the masks along the rows (axis=0) to get the final combined mask
combined_mask = pd.concat(masks, axis=0)

# Use the combined mask to filter the rows in df1 that satisfy the condition (i.e., dates between the date ranges in df2)
dates_in_range_df1 = df1[combined_mask]
```
